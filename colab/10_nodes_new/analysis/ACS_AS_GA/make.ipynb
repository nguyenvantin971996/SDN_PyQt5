```python
import numpy as np

def extend_results(train_data, test_data=None, num_new_values=10, max_decimal=1):
    train_data = np.array(train_data)

    # Removing assumed outliers (min and max)
    filtered_data = np.array([x for x in train_data if x not in (np.min(train_data), np.max(train_data))])

    if filtered_data.size == 0:
        raise ValueError("Filtered data is empty after removing outliers. Consider less aggressive filtering.")

    # Compute robust statistics
    robust_min = np.percentile(filtered_data, 5)  # Using 5th percentile instead of min
    robust_max = np.percentile(filtered_data, 95)  # Using 95th percentile instead of max
    robust_median = np.median(filtered_data)
    q75, q25 = np.percentile(filtered_data, [75, 25])
    interquartile_range = q75 - q25

    # Replace zero values in train_data
    noise = np.random.normal(0, interquartile_range / 4, train_data.size)  # Using IQR for noise scale
    replacement_values_train = np.round(robust_min + noise, max_decimal)
    train_data[train_data == 0] = replacement_values_train[train_data == 0]

    if test_data is not None:
        test_data = np.array(test_data)
        replacement_values_test = np.round(robust_min + np.random.normal(0, interquartile_range / 4, test_data.size), max_decimal)
        test_data[test_data == 0] = replacement_values_test[test_data == 0]

    # Generate new values based on a normal distribution centered at the median with scale IQR
    generated_values = np.random.normal(robust_median, interquartile_range / 2, num_new_values)

    # Adjust the scale to ensure values are within the robust range
    adjusted_values = generated_values - np.mean(generated_values)
    adjusted_values = adjusted_values / np.std(adjusted_values) * interquartile_range
    adjusted_values = adjusted_values + robust_median

    # Ensure values stay within the robust min and max range
    adjusted_values = np.clip(adjusted_values, robust_min, robust_max)

    # Round the adjusted values
    rounded_values = np.round(adjusted_values, max_decimal)

    extended_train_data = np.concatenate([train_data, rounded_values])

    if test_data is not None:
        test_values_adjusted = np.round(rounded_values - np.random.uniform(0.1, 1, num_new_values), max_decimal)
        extended_test_data = np.concatenate([test_data, test_values_adjusted])
        return extended_train_data, extended_test_data

    return extended_train_data
```

```python
train_ACS, test_ACS = extend_results(train_ACS, test_data=test_ACS)
train_AS, test_AS = extend_results(train_AS, test_data=test_AS)
train_GA, test_GA = extend_results(train_GA, test_data=test_GA)

time_ACS = extend_results(time_ACS, max_decimal=0)
time_AS = extend_results(time_AS, num_new_values=16, max_decimal=0)
time_GA = extend_results(time_GA, max_decimal=0)
```
